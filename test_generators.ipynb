```python
import sys
sys.path.append('utils/')

from image_preprocessing import ImageDataGenerator
import matplotlib.pyplot as plt
%matplotlib inline
```

```python
data_generator = ImageDataGenerator(
    rescale=1.0/255.0,
    rotation_range=35, 
    zoom_range=0.2,
    horizontal_flip=True, 
    width_shift_range=0.1,
    height_shift_range=0.1,
    shear_range=0.01,
    channel_shift_range=0.1,
    data_format='channels_last', fill_mode='reflect'
)

train_generator = data_generator.flow_from_directory(
    '/home/dan/data/Caltech-256/train/', 
    target_size=(224, 224),
    batch_size=30
)
```

```python
# get batch
pictures, classes, names = train_generator.next()
```

```python
_, axes = plt.subplots(nrows=6, ncols=5, figsize=(13, 14))
axes = axes.flatten()
for i, pic in enumerate(pictures):
    axes[i].set_axis_off();
    axes[i].imshow(pic);
    axes[i].set_title(names[i]);
plt.tight_layout()
```
